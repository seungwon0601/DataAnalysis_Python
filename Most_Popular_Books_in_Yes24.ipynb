# What are the most popular books in Yes24
- Scrap data from [yes24.com](http://www.yes24.com/Main/default.aspx)
- Keyword = 'Python'
- Volumn = '10 pages | 20 books per 1 page'


```
from bs4 import BeautifulSoup
import requests
```

```
test_yes24_python = requests.get('http://www.yes24.com/searchcorner/Search?keywordAd=&keyword=&domain=BOOK&qdomain=%B1%B9%B3%BB%B5%B5%BC%AD&query=%C6%C4%C0%CC%BD%E3&PageNumber=1&scode=012')
soup = BeautifulSoup(test_yes24_python.text, 'html.parser')
```

### BeutifulSoup을 통해 가져온 html을 출력해봅니다.

```
print(soup.prettify())
```

### (test)select 문법을 사용해서 하나의 페이지 내 도서 리스트만 추출해봅니다.

```
test = soup.select('tr a > strong') # 'tr > a > strong' 이 아님. '>'는 바로 하위 태그인 경우만 사용
print(test)
print('------------------------------')
for i in range(len(test)):
    print(test[i].string)
```

### 이상없다면, 총 10페이지 약 200권 분량의 도서 목록과 판매지수를 추출해봅니다.

```
book_list = []
score_list = []
for i in range(1, 10): # 1~10페이지까지 인기순으로 모든 도서 출력
    yes24_python = requests.get('http://www.yes24.com/searchcorner/Search?keywordAd=&keyword=&domain=BOOK&qdomain=%B1%B9%B3%BB%B5%B5%BC%AD&query=%C6%C4%C0%CC%BD%E3&PageNumber={}&scode=012'.format(i))
    soup = BeautifulSoup(yes24_python.text, 'html.parser')
    #도서명 추출
    books = soup.select('tr a > strong')
    for j in range(len(books)):
        print(books[j].string)
        book_list.append(books[j].string)

    #판매지수 추출
    scores = soup.select('tr div.goods_rating > span.gd_reviewCount')
    for k in range(len(scores)):
        print(scores[k].text.strip())
        score_list.append(scores[k].text.strip())
```

### 데이터프레임 생성을 위해 추출된 도서와 판매지수 리스트 길이가 일치하는지 확인합니다.
